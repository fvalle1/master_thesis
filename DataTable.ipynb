```python
import os
import sys
import pandas as pd
import numpy as np
import regex as re
from matplotlib import pyplot as plt
```

```python
working_dir = "/Users/filippo/Developer/tesi"
os.chdir(working_dir)
dirs = os.listdir("data")
```

```python
df = pd.read_csv(("%s/genes.txt"%working_dir), header=None)
df.columns=["gene"]
print(len(df['gene']))
df.head()
```

```python
def getFilenameFromDir(directory):
    for element in os.listdir(directory):
        if re.match("[a-zA-Z0-9]{8}-[a-zA-Z0-9]{4}-[a-zA-Z0-9]{4}-[a-zA-Z0-9\-]{4}-[a-zA-Z0-9\-]{12}[\.FPKM]{5}.txt[\.gz]{0,3}",element):
            cfile = element
            print(element)
            return cfile
```

```python
maxacceptables = 100
for i,cdirectory in enumerate(dirs):
    cfile = getFilenameFromDir("/Users/filippo/Developer/tesi/data/%s"%cdirectory)
    cdf = pd.read_csv(("%s/data/%s/%s"%(working_dir,cdirectory,cfile)), sep='\t', header=None)
    cdf.columns = ["gene", cfile[:-5]]
    try:
        df.insert(1,cdf.keys()[1][:-5],cdf.loc[:,cdf.keys()[1]].values)
    except:
        print "Unexpected error:", sys.exc_info()[0]
    if i > maxacceptables:
        break
```

```python
A = [df.loc[:,df.keys()[89]].values, df.loc[:,df.keys()[45]].values, df.loc[:,df.keys()[42]].values]
colors = ['red', 'blue', 'yellow']
fig = plt.figure()
x = np.arange(1, len(A[0]))
plt.ylabel("frequencyi")
plt.xlabel("RANKi")
plt.yscale('log')
plt.xscale('log')
plt.plot(x, 1./x, 'g--')
for i,a in enumerate(A):
    plt.plot(np.sort(a/np.sum(a))[::-1], c=colors[i]) 
plt.show()
```
